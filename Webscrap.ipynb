```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
```

```python
uClient =uReq('https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_0_3_na_na_pr&otracker1=AS_QueryStore_OrganicAutoSuggest_0_3_na_na_pr&as-pos=0&as-type=RECENT&suggestionId=laptop&requestId=3af7f76b-cfdc-490d-a3cb-3dec368ee49d&as-backfill=on'+'&page=1')
```

```python
page_html=uClient.read()
```

```python
page_html
```

```python
page_soup=soup(page_html,"html.parser")
```

```python
noOfCust=page_soup.findAll("span",{"class": "_38sUEc"})
```

```python
noOfCust[0].text.split('Ratings')[0].replace(',','').strip()
```
